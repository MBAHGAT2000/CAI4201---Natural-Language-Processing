```python
corpus = ["this is a sentence is","this is another sentence","third document is here"]
result = []
vocab = {}
for sent in corpus:
    for word in sent.split():
        if word not in result and len(word)>1:
            result.append(word)
result = sorted(result)
for index in range(len(result)):
    vocab[index] = result[index]
```

# Vectorization
- Vectorization : process of encoding text as integers to create feature vecors
- Feature vector : vector of numerical features that represent an object
- Types Of Vecorization (count vectorization - Ngrams - TFIDF)
count vectorization  == count unique words occure in the sentence how many times!! 

# Bag of words ( Count Vecorization )

- Count the occurrences of words in the corpus.

```python
# create a document term matrix
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
vectorizer = CountVectorizer()
```

###### apply on a simple corpus

```python
corpus = ["this is a sentence is",
          "this is another sentence",
          "third document is here"]

X = vectorizer.fit(corpus)
```

```python
print(X.vocabulary_) # automatically remove any one character word like a
print(vectorizer.get_feature_names_out()) # unique words in corpus
```

```python
X = vectorizer.transform(corpus) # create a document term matrix
print(X.shape)
print(X.toarray()) #document term matrix
```

```python
columns = vectorizer.get_feature_names_out()
pd.DataFrame(X.toarray(), columns=columns, index=corpus)
```

# 2- N-Grams Vecorization

N-Grams is a way we can use to count for the context in the text, the bigger n-gram range the bigger context you can capture but also more features to generate, so be careful not to break your memory.

### N-Grams create document term matrix
- columns represent all columns of adjacent words of length n
- cells represent count
Example : "i am studying nlp"
| Character  | function | sentences |
| ------------- | ------------- | -------- |
|bigram |  ["i am"  ,  "am studying"  ,   "studying nlp"] |  3 sentences |
|trigram |  ["i am studying" , "am studying nlp"]|  2 sentences |
|4-gram  |  ["i am studying nlp"]|  1 sentences |

### BiGram only

```python
vectorizer_bigram = CountVectorizer(ngram_range=(2,2))
texts = ["this is a sentence is a sentence", "this is another sentence", "third document is here"]
```

```python
vectorizer_bigram.fit(texts)
x = vectorizer_bigram.transform(texts)
columns = vectorizer_bigram.get_feature_names_out()
pd.DataFrame(x.toarray(), columns=columns, index=texts)
```

### TriGram only

```python
vectorizer_trigram = CountVectorizer(ngram_range=(3,3))
vectorizer_trigram.fit(texts)
x = vectorizer_trigram.transform(texts)
columns = vectorizer_trigram.get_feature_names_out()
pd.DataFrame(x.toarray(), columns=columns, index=texts)
```

## mean bigram and trigram

# 3- TF-IDF Vecorization
- Instead of just counting the frequency of each word, each word here is weighted using TF-IDF

$$W_{x, y} = tf_{x, y} \times log(\frac{N}{df_x})$$

| Terms  | results  |
| ------------- | ------------- | 
|TF-IDF |create document term matrix|
|columns |  are individual unique words |  
|cells |  contain a weight which signifies how important a word is for an individual text message W(i,j) = TF(i,j) * log(N/DF(i)) | 
| N  | total number of documents  |  
|TF(i,j)  |number of times word i is occured in sentence j divided by total number of terms in sentence j |
|DF(i) |number of documents containing word i|
|j|"i am studying nlp"|
|TF(am,j)= 1/4 = 0.25 | occure only one time and the sentence has 4 words|
|N|200|
| DF(am) |2|
| W(am,j)| TF(am,j) * log(N/DF(am)) = 0.25 * log(200/2) = 0.5|

```python
from sklearn.feature_extraction.text import TfidfVectorizer
corpus = ["this is a sentence is", "this is another sentence", "third document is here"]
'''
this is a sentence is
tf       this 1/3         is 2/3   sentence 1/3
idf      this 1.1249    is 1   sentence 1.1249
tf-idf   this 0.3749    is 2/3   sentence 0.374967
L2 = 0.8518
'''
tfidf_vec = TfidfVectorizer()
tfidf_vec.fit(corpus)
x = tfidf_vec.transform(corpus)
columns = tfidf_vec.get_feature_names_out()
pd.DataFrame(x.toarray(), columns=columns, index=corpus)
```

```python
# fit transform directly
from sklearn.feature_extraction.text import TfidfVectorizer
corpus = ["this is a sentence is", "this is another sentence", "third document is here"]
tfidf_vec = TfidfVectorizer()
x = tfidf_vec.fit_transform(corpus)
columns = tfidf_vec.get_feature_names_out()
pd.DataFrame(x.toarray(), columns=columns, index=corpus)
```
